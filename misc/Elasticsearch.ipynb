# Enriched - github_issues_chaoss_enriched (by github Enricher) 

```python
search = scan(
    es,
    index='github_issues_chaoss_enriched',
    query={"query": { "match_all" : {}}}
)
hits = []
for hit in search:
    hits.append(hit)
df_enriched_github_issues = json_normalize(hits)
df_enriched_github_issues.shape[0]
```

```python
df_enriched_github_issues.head()
```

```python
df_enriched_github_issues.shape[1]
```

```python
compression_opts = dict(method='zip',
                        archive_name='out2.csv')  
df_enriched_github_issues.to_csv('out2.zip', index=False,
          compression=compression_opts) 
```
